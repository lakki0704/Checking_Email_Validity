```python
def validate_emails(id):
    import requests
    api_key = "ae68007f-6ee7-4616-a0de-0c39bb50f31e" 
    email_address = id
    response =requests.get("https://isitarealemail.com/api/email/validate", params = {'email': email_address}, headers = {'Authorization': "Bearer " + api_key })
    status = response.json()['status']
    if status == "valid":
        print("email is valid")
    elif status == "invalid":
        print("email is invalid")
    else:
        print("email was unknown")
    
import pandas as pd
import csv
# Takes the file's folder
filepath = r"C:\Users\laksh\Downloads\wrongemail.csv"
   
# read the CSV file
df = pd.read_csv(filepath,usecols = ['Emails'])


for i in range(1787):
    print(i,df.loc[i].to_string())
    validate_emails(df.loc[i])
    print("\n")

```
